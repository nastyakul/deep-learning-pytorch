```python
import numpy as np
```

```python
# Write a function that takes as input two lists Y, P,
# and returns the float corresponding to their cross-entropy.
def cross_entropy(Y, P):
    return -np.sum(Y * np.log(P) + (np.ones(len(Y))-Y) * np.log(np.ones(len(Y)) - P))
```

```python
Y = [0, 0, 1]
P = [0.8, 0.7, 0.1]
```

```python
cross_entropy(Y, P)
```
